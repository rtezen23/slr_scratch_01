```python
# importing libraries and data
import pandas as pd
import numpy as np

df = pd.read_csv('../data/Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Elegimos aleatoriamente 300 índices para la columna Age y 54 para Gender
age_nan_indices = np.random.choice(df.index, size=300, replace=False)
df.loc[age_nan_indices, 'Age'] = np.nan  # Asignamos NaN en esos índices

gender_nan_indices = np.random.choice(df.index, size=54, replace=False)
df.loc[gender_nan_indices, 'Gender'] = np.nan  # Asignamos NaN en esos índices
```

```python
df.info()
```

```python
df_drop_na = df.dropna(axis=1)
```

```python
df_drop_na.info()
```

```python
df_drop_na_rows = df.dropna()
```

```python
df_drop_na_rows.info()
```

```python
df_fillna = df.copy()
```

```python
df_fillna['Gender'].mode()
```

```python
df_fillna = df_fillna.fillna({
    'Gender': df_fillna['Gender'].mode()[0],
    'Age': df_fillna['Age'].mean(),
})
```

```python
df_fillna.info()
```

```python
df_bfill = df.copy()
```

```python
df_bfill.info()
```

```python
df_bfill['Age'] = df_bfill['Age'].bfill()
df_bfill.info()
```
